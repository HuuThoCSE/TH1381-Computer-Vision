```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('D:\\TH1381-Computer-Vision\\OnThi\\Thamkhao_Old\\Tuan3\\dog_backpack.png')
img2 = cv2.imread('D:\\TH1381-Computer-Vision\\OnThi\\Thamkhao_Old\\Tuan3\\watermark_no_copy.png')
img2 = cv2.resize(img2, (600, 600))

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
img2gray.shape
```

```python
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
mask_inv.shape
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
x_offset = 934-600
y_offset = 1401-600
```

```python
rows, cols, channels = img2.shape
roi = img1[y_offset:1401, x_offset:943]
```

```python
roi.shape
```

```python
plt.imshow(roi)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
fg.shape
```

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img
```

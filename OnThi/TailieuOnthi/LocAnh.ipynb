```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_url_dog = 'D:\\Github\\TH1381-Computer-Vision\\OnThi\\TailieuOnthi\\image\\dog.jpg'
```

```python
# Đọc ảnh
image_path = img_url_dog
image = cv2.imread(image_path)
```

```python
# Chuyển ảnh sang ảnh xám
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Hiển thị kết quả
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Ảnh gốc')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.title('Ảnh xám')
plt.axis('off')
```

```python
## Bộ lọc trung bình (box blur)
kernel_size = 5
box_blur_kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
```

```python
## Tạo bộ lọc làm nét
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]], np.float32)
```

```python
# Thực hiện lọc làm mịn (box blur)
box_blur_image = cv2.filter2D(gray_image, -1, box_blur_kernel)
```

```python
# Thực hiện lọc làm nét
sharpened_image = cv2.filter2D(gray_image, -1, sharpen_kernel)
```

```python
# Hiển thị kết quả
plt.figure(figsize=(15, 5))

# Ảnh gốc
plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh gốc')
plt.axis('off')

# Ảnh sau khi lọc làm mịn (box blur)
plt.subplot(1, 3, 2)
plt.imshow(box_blur_image, cmap='gray')
plt.title('Lọc làm mịn (Box Blur)')
plt.axis('off')

# Ảnh sau khi lọc làm nét
plt.subplot(1, 3, 3)
plt.imshow(sharpened_image, cmap='gray')
plt.title('Lọc làm nét')
plt.axis('off')

plt.tight_layout()
plt.show()
```

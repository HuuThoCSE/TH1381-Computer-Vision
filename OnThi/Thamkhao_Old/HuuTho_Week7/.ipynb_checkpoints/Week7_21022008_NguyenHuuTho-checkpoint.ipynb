Biến đổi Hough (Hough Transform) là một kỹ thuật trong thị giác máy tính được sử dụng để phát hiện các đối tượng có hình dạng cụ thể trong hình ảnh, ví dụ như đường thẳng, hình tròn, hình elip, v.v. Dưới đây là một giải thích dễ hiểu về cách hoạt động của biến đổi Hough:

### 1. Mục đích
Biến đổi Hough được sử dụng để xác định các đối tượng có hình dạng xác định bằng cách biến đổi từ không gian ảnh (image space) sang không gian tham số (parameter space).

### 2. Nguyên lý cơ bản
Giả sử ta muốn phát hiện các đường thẳng trong một hình ảnh. Mỗi đường thẳng trong không gian ảnh có thể được biểu diễn bằng phương trình:
\[ y = ax + b \]
hoặc trong hệ tọa độ cực:
\[ r = x \cos \theta + y \sin \theta \]
Trong đó:
- \( r \) là khoảng cách từ gốc tọa độ đến đường thẳng.
- \( \theta \) là góc giữa đoạn thẳng nối gốc tọa độ và đường thẳng với trục hoành.

### 3. Không gian tham số
Mỗi điểm trong không gian ảnh tương ứng với một đường cong trong không gian tham số. Khi nhiều điểm nằm trên cùng một đường thẳng trong không gian ảnh, các đường cong tương ứng trong không gian tham số sẽ giao nhau tại cùng một điểm. Điểm giao nhau này đại diện cho các tham số \( r \) và \( \theta \) của đường thẳng đó.

### 4. Các bước thực hiện
1. **Phát hiện các điểm biên (edge detection):** Sử dụng các kỹ thuật như bộ lọc Canny để tìm các điểm biên trong hình ảnh.
2. **Biến đổi Hough:** Mỗi điểm biên sẽ tạo ra một đường cong trong không gian tham số. Lưu các điểm giao nhau của các đường cong này trong một mảng số đếm.
3. **Phát hiện các đỉnh (peak detection):** Tìm các điểm có số đếm cao nhất trong không gian tham số. Các điểm này tương ứng với các đường thẳng trong không gian ảnh.

### 5. Ứng dụng
- **Phát hiện đường thẳng:** Để tìm các cạnh của đối tượng trong hình ảnh, chẳng hạn như các làn đường trong hệ thống lái tự động.
- **Phát hiện hình tròn:** Để nhận diện các đối tượng hình tròn như đồng xu, bóng đèn, hoặc các nắp chai.

### Ví dụ minh họa
Giả sử ta có một hình ảnh chứa các đường thẳng và muốn phát hiện chúng:

1. **Hình ảnh gốc với các đường thẳng:**

2. **Biên phát hiện sau khi áp dụng bộ lọc Canny:**

3. **Không gian tham số sau khi áp dụng biến đổi Hough:**

4. **Các đường thẳng được phát hiện:**

Qua các bước trên, ta có thể dễ dàng xác định và đánh dấu các đường thẳng trong hình ảnh, phục vụ cho nhiều ứng dụng trong thị giác máy tính và phân tích hình ảnh.

Trong OpenCV, có hai hàm chính để phát hiện đường thẳng dựa trên biến đổi Hough là `HoughLines` và `HoughLinesP`. Dưới đây là cú pháp và cách sử dụng của hai hàm này:

### 1. `HoughLines`
Hàm này được sử dụng để phát hiện các đường thẳng trong không gian cực (polar space).

**Cú pháp:**
```python
lines = cv2.HoughLines(image, rho, theta, threshold)
```

**Tham số:**
- `image`: Ảnh đầu vào, phải là ảnh nhị phân (binary image), thường là kết quả của bộ lọc Canny.
- `rho`: Độ phân giải của tham số \( \rho \) trong không gian cực. Thông thường là 1 pixel.
- `theta`: Độ phân giải của tham số \( \theta \) trong không gian cực. Thông thường là `np.pi/180` (1 độ).
- `threshold`: Ngưỡng số đếm. Chỉ những đường thẳng có số điểm lớn hơn ngưỡng này mới được xem xét.

**Ví dụ:**
```python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('path_to_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Áp dụng bộ lọc Canny
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Sử dụng biến đổi Hough để phát hiện đường thẳng
lines = cv2.HoughLines(edges, 1, np.pi/180, 200)

# Vẽ các đường thẳng lên ảnh gốc
for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

cv2.imshow('Detected Lines', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 2. `HoughLinesP`
Hàm này là một phiên bản của `HoughLines` nhưng sử dụng biến đổi Hough xác suất (probabilistic Hough transform), giúp phát hiện các đoạn thẳng (line segments) thay vì các đường thẳng vô hạn.

**Cú pháp:**
```python
lines = cv2.HoughLinesP(image, rho, theta, threshold, minLineLength=None, maxLineGap=None)
```

**Tham số:**
- `image`: Ảnh đầu vào, phải là ảnh nhị phân (binary image).
- `rho`: Độ phân giải của tham số \( \rho \) trong không gian cực.
- `theta`: Độ phân giải của tham số \( \theta \) trong không gian cực.
- `threshold`: Ngưỡng số đếm.
- `minLineLength`: Chiều dài tối thiểu của đoạn thẳng. Các đoạn thẳng ngắn hơn giá trị này sẽ bị bỏ qua.
- `maxLineGap`: Khoảng cách tối đa giữa hai đoạn thẳng gần kề để chúng được nối lại thành một đoạn thẳng.

**Ví dụ:**
```python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('path_to_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Áp dụng bộ lọc Canny
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Sử dụng biến đổi Hough xác suất để phát hiện đoạn thẳng
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, minLineLength=50, maxLineGap=10)

# Vẽ các đoạn thẳng lên ảnh gốc
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

cv2.imshow('Detected Line Segments', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Cả hai hàm này đều giúp phát hiện các đường thẳng trong ảnh, nhưng `HoughLinesP` thường được ưa chuộng hơn do khả năng phát hiện các đoạn thẳng cụ thể và kiểm soát tốt hơn về chiều dài và khoảng cách giữa các đoạn thẳng.

````python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('/content/drive/MyDrive/23_24_HK2/CoVi/CoVi_Pactive/Week7/sudoku5.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Áp dụng bộ lọc Canny
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Sử dụng biến đổi Hough xác suất để phát hiện đoạn thẳng
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, minLineLength=50, maxLineGap=10)

# Vẽ các đoạn thẳng lên ảnh gốc
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Trong OpenCV, hàm được sử dụng để phát hiện các đường tròn trong ảnh là `HoughCircles`. Dưới đây là cú pháp và cách sử dụng của hàm này:

### `HoughCircles`
Hàm này được sử dụng để phát hiện các đường tròn trong ảnh bằng cách sử dụng biến đổi Hough.

**Cú pháp:**
```python
circles = cv2.HoughCircles(image, method, dp, minDist, param1=None, param2=None, minRadius=None, maxRadius=None)
```

**Tham số:**
- `image`: Ảnh đầu vào, phải là ảnh xám (grayscale image).
- `method`: Phương pháp biến đổi Hough được sử dụng. Thường là `cv2.HOUGH_GRADIENT`.
- `dp`: Tỉ lệ nghịch của độ phân giải bộ tích lũy. Nếu `dp=1`, độ phân giải của bộ tích lũy giống như độ phân giải của ảnh đầu vào. Nếu `dp=2`, độ phân giải của bộ tích lũy chỉ bằng một nửa độ phân giải của ảnh đầu vào.
- `minDist`: Khoảng cách tâm tối thiểu giữa các đường tròn được phát hiện.
- `param1`: Ngưỡng cao cho bộ lọc Canny (cạnh trên cho phát hiện cạnh).
- `param2`: Ngưỡng cho bộ tích lũy trung bình Hough. Giá trị này càng nhỏ thì càng có nhiều đường tròn được phát hiện (bao gồm cả các đường tròn giả).
- `minRadius`: Bán kính nhỏ nhất của đường tròn cần phát hiện.
- `maxRadius`: Bán kính lớn nhất của đường tròn cần phát hiện.

**Ví dụ:**
```python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('path_to_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Áp dụng bộ lọc Gaussian để làm mờ ảnh và giảm nhiễu
gray = cv2.GaussianBlur(gray, (9, 9), 2)

# Sử dụng biến đổi Hough để phát hiện các đường tròn
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2, minDist=20,
                           param1=50, param2=30, minRadius=10, maxRadius=100)

# Nếu có ít nhất một đường tròn được phát hiện
if circles is not None:
    # Chuyển các tham số của đường tròn từ float sang int
    circles = np.round(circles[0, :]).astype("int")

    # Vẽ các đường tròn được phát hiện lên ảnh gốc
    for (x, y, r) in circles:
        cv2.circle(image, (x, y), r, (0, 255, 0), 2)
        cv2.circle(image, (x, y), 2, (0, 0, 255), 3)  # Vẽ tâm của đường tròn

# Hiển thị ảnh kết quả
cv2.imshow('Detected Circles', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Giải thích chi tiết ví dụ:
1. **Đọc và chuyển đổi ảnh sang ảnh xám:**
   - `cv2.imread('path_to_image.jpg')`: Đọc ảnh từ tệp.
   - `cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)`: Chuyển đổi ảnh màu sang ảnh xám.

2. **Làm mờ ảnh:**
   - `cv2.GaussianBlur(gray, (9, 9), 2)`: Áp dụng bộ lọc Gaussian để làm mờ ảnh và giảm nhiễu.

3. **Phát hiện các đường tròn:**
   - `cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2, minDist=20, param1=50, param2=30, minRadius=10, maxRadius=100)`: Sử dụng biến đổi Hough để phát hiện các đường tròn.

4. **Vẽ các đường tròn lên ảnh gốc:**
   - `cv2.circle(image, (x, y), r, (0, 255, 0), 2)`: Vẽ đường tròn phát hiện được.
   - `cv2.circle(image, (x, y), 2, (0, 0, 255), 3)`: Vẽ tâm của đường tròn.

5. **Hiển thị ảnh kết quả:**
   - `cv2.imshow('Detected Circles', image)`: Hiển thị ảnh có các đường tròn được phát hiện.
   - `cv2.waitKey(0)`: Đợi một phím bất kỳ để đóng cửa sổ hiển thị.
   - `cv2.destroyAllWindows()`: Đóng tất cả các cửa sổ hiển thị.

Với cú pháp và ví dụ trên, bạn có thể dễ dàng phát hiện và vẽ các đường tròn trong ảnh sử dụng OpenCV.

````python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('/content/drive/MyDrive/23_24_HK2/CoVi/CoVi_Pactive/Week7/whitebloodcell1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Áp dụng bộ lọc Gaussian để làm mờ ảnh và giảm nhiễu
gray = cv2.GaussianBlur(gray, (9, 9), 2)

# Sử dụng biến đổi Hough để phát hiện các đường tròn
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2, minDist=20,
                           param1=50, param2=30, minRadius=10, maxRadius=100)

# Nếu có ít nhất một đường tròn được phát hiện
if circles is not None:
    # Chuyển các tham số của đường tròn từ float sang int
    circles = np.round(circles[0, :]).astype("int")

    # Vẽ các đường tròn được phát hiện lên ảnh gốc
    for (x, y, r) in circles:
        cv2.circle(image, (x, y), r, (0, 255, 0), 2)
        cv2.circle(image, (x, y), 2, (0, 0, 255), 3)  # Vẽ tâm của đường tròn

# Hiển thị ảnh kết quả
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

# **PHÁT HIỆN GÓC (CORNER DETECTION)**

Giải thuật phát hiện góc Harris là một trong những phương pháp phổ biến để phát hiện các điểm góc (corner points) trong hình ảnh. Góc là các điểm trong hình ảnh nơi cường độ sáng thay đổi đáng kể theo mọi hướng. Đây là những điểm quan trọng trong nhiều ứng dụng thị giác máy tính như nhận diện đối tượng, theo dõi chuyển động, và ghép ảnh.

### Giải thích dễ hiểu về giải thuật phát hiện góc Harris

#### 1. Khái niệm cơ bản
Góc là nơi cường độ pixel thay đổi lớn trong nhiều hướng khác nhau. Giải thuật phát hiện góc Harris hoạt động bằng cách đo lường sự thay đổi cường độ khi di chuyển cửa sổ nhỏ quanh điểm đó.

#### 2. Các bước chính của giải thuật
1. **Tính đạo hàm bậc nhất của ảnh:**
   - Tính đạo hàm theo hướng x (\(I_x\)) và đạo hàm theo hướng y (\(I_y\)) của ảnh để tìm sự thay đổi cường độ theo hai hướng này.

2. **Tính ma trận cấu trúc (structure tensor):**
   - Ma trận cấu trúc tại mỗi điểm ảnh \((x, y)\) được tính như sau:
     \[
     M = \begin{bmatrix}
     I_x^2 & I_x I_y \\
     I_x I_y & I_y^2
     \end{bmatrix}
     \]
   - Trong đó, các giá trị trong ma trận là tổng của các tích phân các giá trị đạo hàm trong một cửa sổ nhỏ xung quanh điểm \((x, y)\).

3. **Tính thước đo góc (corner response function):**
   - Thước đo góc được tính bằng cách sử dụng ma trận cấu trúc:
     \[
     R = \text{det}(M) - k \cdot (\text{trace}(M))^2
     \]
   - Trong đó, \(\text{det}(M) = I_x^2 I_y^2 - (I_x I_y)^2\) là định thức của ma trận M, \(\text{trace}(M) = I_x^2 + I_y^2\) là tổng của các phần tử trên đường chéo chính của ma trận M, và \(k\) là một hằng số nhỏ thường được chọn trong khoảng từ 0.04 đến 0.06.

4. **Ngưỡng hóa thước đo góc:**
   - Các điểm có giá trị R lớn hơn một ngưỡng nhất định được xem là các điểm góc.

5. **Non-maximum suppression (NMS):**
   - Để loại bỏ các điểm góc không phải là cực đại địa phương, non-maximum suppression được áp dụng. Điều này đảm bảo chỉ giữ lại các điểm góc thực sự nổi bật.

#### 3. Cú pháp hàm Harris trong OpenCV

Trong OpenCV, hàm `cv2.cornerHarris` được sử dụng để thực hiện phát hiện góc Harris.

**Cú pháp:**
```python
dst = cv2.cornerHarris(src, blockSize, ksize, k)
```

**Tham số:**
- `src`: Ảnh đầu vào, phải là ảnh xám (grayscale image).
- `blockSize`: Kích thước của cửa sổ được sử dụng để tính đạo hàm.
- `ksize`: Kích thước của bộ lọc Sobel được sử dụng để tính đạo hàm ảnh.
- `k`: Hằng số tự do trong phương trình thước đo góc Harris.

**Ví dụ:**
```python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('path_to_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Phát hiện góc Harris
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Kết quả là ảnh có các giá trị góc, áp dụng ngưỡng để đánh dấu các góc
dst = cv2.dilate(dst, None)
image[dst > 0.01 * dst.max()] = [0, 0, 255]

# Hiển thị ảnh kết quả
cv2.imshow('Detected Corners', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Giải thích chi tiết ví dụ:
1. **Đọc và chuyển đổi ảnh sang ảnh xám:**
   - `cv2.imread('path_to_image.jpg')`: Đọc ảnh từ tệp.
   - `cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)`: Chuyển đổi ảnh màu sang ảnh xám.

2. **Phát hiện góc Harris:**
   - `np.float32(gray)`: Chuyển đổi ảnh xám sang kiểu dữ liệu float32 để phù hợp với hàm Harris.
   - `cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)`: Sử dụng biến đổi Harris để phát hiện các góc trong ảnh.

3. **Đánh dấu các điểm góc:**
   - `cv2.dilate(dst, None)`: Làm dày các điểm góc để dễ dàng nhìn thấy.
   - `image[dst > 0.01 * dst.max()] = [0, 0, 255]`: Đánh dấu các điểm góc trên ảnh gốc bằng màu đỏ.

4. **Hiển thị ảnh kết quả:**
   - `cv2.imshow('Detected Corners', image)`: Hiển thị ảnh có các điểm góc được đánh dấu.
   - `cv2.waitKey(0)`: Đợi một phím bất kỳ để đóng cửa sổ hiển thị.
   - `cv2.destroyAllWindows()`: Đóng tất cả các cửa sổ hiển thị.

Với cú pháp và ví dụ trên, bạn có thể dễ dàng phát hiện và đánh dấu các điểm góc trong ảnh sử dụng OpenCV.

````python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('/content/drive/MyDrive/23_24_HK2/CoVi/CoVi_Pactive/Week7/puppy2.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Phát hiện góc Harris
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Kết quả là ảnh có các giá trị góc, áp dụng ngưỡng để đánh dấu các góc
dst = cv2.dilate(dst, None)
image[dst > 0.01 * dst.max()] = [0, 0, 255]

# Hiển thị ảnh kết quả
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Giải thuật này dựa trên tiêu chí đơn giản hơn và chính xác hơn so với Harris, giúp phát hiện các điểm góc có chất lượng tốt hơn cho các ứng dụng như theo dõi đối tượng và ghép ảnh.

### Giải thích về giải thuật Shi-Tomasi

#### 1. Khái niệm cơ bản
Giải thuật Shi-Tomasi cũng sử dụng ma trận cấu trúc (structure tensor) giống như giải thuật Harris. Tuy nhiên, thay vì sử dụng một thước đo góc phức tạp, Shi-Tomasi đơn giản hóa bằng cách chọn ra giá trị nhỏ nhất trong hai giá trị eigenvalue của ma trận cấu trúc.

#### 2. Các bước chính của giải thuật
1. **Tính đạo hàm bậc nhất của ảnh:**
   - Tính đạo hàm theo hướng x (\(I_x\)) và đạo hàm theo hướng y (\(I_y\)) của ảnh để tìm sự thay đổi cường độ theo hai hướng này.

2. **Tính ma trận cấu trúc (structure tensor):**
   - Ma trận cấu trúc tại mỗi điểm ảnh \((x, y)\) được tính như sau:
     \[
     M = \begin{bmatrix}
     I_x^2 & I_x I_y \\
     I_x I_y & I_y^2
     \end{bmatrix}
     \]

3. **Tính giá trị eigenvalue:**
   - Tính hai giá trị eigenvalue \(\lambda_1\) và \(\lambda_2\) của ma trận \(M\).

4. **Thước đo góc Shi-Tomasi:**
   - Sử dụng giá trị nhỏ nhất của hai eigenvalue làm thước đo góc:
     \[
     R = \min(\lambda_1, \lambda_2)
     \]

5. **Ngưỡng hóa thước đo góc:**
   - Các điểm có giá trị \(R\) lớn hơn một ngưỡng nhất định được xem là các điểm góc.

6. **Non-maximum suppression (NMS):**
   - Để loại bỏ các điểm góc không phải là cực đại địa phương, non-maximum suppression được áp dụng. Điều này đảm bảo chỉ giữ lại các điểm góc thực sự nổi bật.

### Cú pháp hàm Shi-Tomasi trong OpenCV

Trong OpenCV, hàm `cv2.goodFeaturesToTrack` được sử dụng để phát hiện các góc Shi-Tomasi.

**Cú pháp:**
```python
corners = cv2.goodFeaturesToTrack(image, maxCorners, qualityLevel, minDistance, mask=None, blockSize=None, useHarrisDetector=None, k=None)
```

**Tham số:**
- `image`: Ảnh đầu vào, phải là ảnh xám (grayscale image).
- `maxCorners`: Số lượng góc tối đa cần phát hiện. Nếu là `0`, không giới hạn số lượng góc.
- `qualityLevel`: Giá trị biểu thị chất lượng tối thiểu của góc để được chấp nhận (giá trị giữa 0 và 1).
- `minDistance`: Khoảng cách Euclidean nhỏ nhất giữa các góc được phát hiện.
- `mask`: Ảnh mặt nạ (mask image) để chỉ định khu vực cần phát hiện góc.
- `blockSize`: Kích thước của cửa sổ được sử dụng để tính đạo hàm.
- `useHarrisDetector`: Có sử dụng thước đo góc Harris hay không (Boolean).
- `k`: Tham số tự do cho thước đo góc Harris (nếu `useHarrisDetector=True`).

### Ví dụ minh họa

```python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('path_to_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Sử dụng giải thuật Shi-Tomasi để phát hiện các điểm góc
corners = cv2.goodFeaturesToTrack(gray, maxCorners=100, qualityLevel=0.01, minDistance=10, blockSize=3)

# Chuyển các điểm góc từ float sang int
corners = np.int0(corners)

# Vẽ các điểm góc lên ảnh gốc
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(image, (x, y), 3, (0, 255, 0), -1)

# Hiển thị ảnh kết quả
cv2.imshow('Detected Corners', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Giải thích chi tiết ví dụ:
1. **Đọc và chuyển đổi ảnh sang ảnh xám:**
   - `cv2.imread('path_to_image.jpg')`: Đọc ảnh từ tệp.
   - `cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)`: Chuyển đổi ảnh màu sang ảnh xám.

2. **Phát hiện góc Shi-Tomasi:**
   - `cv2.goodFeaturesToTrack(gray, maxCorners=100, qualityLevel=0.01, minDistance=10, blockSize=3)`: Sử dụng giải thuật Shi-Tomasi để phát hiện các điểm góc trong ảnh.
     - `maxCorners=100`: Chỉ phát hiện tối đa 100 góc.
     - `qualityLevel=0.01`: Các góc có chất lượng lớn hơn 1% giá trị góc lớn nhất sẽ được giữ lại.
     - `minDistance=10`: Các góc được phát hiện phải cách nhau ít nhất 10 pixel.
     - `blockSize=3`: Kích thước của cửa sổ được sử dụng để tính đạo hàm.

3. **Chuyển đổi và vẽ các điểm góc:**
   - `corners = np.int0(corners)`: Chuyển đổi các tọa độ góc từ float sang int.
   - `cv2.circle(image, (x, y), 3, (0, 255, 0), -1)`: Vẽ các điểm góc lên ảnh gốc bằng các vòng tròn màu xanh lá cây.

4. **Hiển thị ảnh kết quả:**
   - `cv2.imshow('Detected Corners', image)`: Hiển thị ảnh có các điểm góc được đánh dấu.
   - `cv2.waitKey(0)`: Đợi một phím bất kỳ để đóng cửa sổ hiển thị.
   - `cv2.destroyAllWindows()`: Đóng tất cả các cửa sổ hiển thị.

Với cú pháp và ví dụ trên, bạn có thể dễ dàng phát hiện và đánh dấu các điểm góc trong ảnh sử dụng giải thuật Shi-Tomasi trong OpenCV.

````python
import cv2
import numpy as np

# Đọc ảnh và chuyển sang ảnh xám
image = cv2.imread('/content/drive/MyDrive/23_24_HK2/CoVi/CoVi_Pactive/Week7/Dog.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Sử dụng giải thuật Shi-Tomasi để phát hiện các điểm góc
corners = cv2.goodFeaturesToTrack(gray, maxCorners=100, qualityLevel=0.01, minDistance=10, blockSize=3)

# Chuyển các điểm góc từ float sang int
corners = np.int0(corners)

# Vẽ các điểm góc lên ảnh gốc
for corner in corners:
    x, y = corner.ravel()
    cv2.circle(image, (x, y), 3, (0, 255, 0), -1)

# Hiển thị ảnh kết quả
cv2_imshow(image)
````

### M ngun minh ha SIFT
```python
import cv2
import matplotlib.pyplot as plt

# c hnh nh
image = cv2.imread('duong_dan_toi_anh.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Khi to SIFT detector
sift = cv2.SIFT_create()

# Tm key points v descriptors vi SIFT
keypoints, descriptors = sift.detectAndCompute(gray_image, None)

# V key points ln hnh nh
img_with_keypoints = cv2.drawKeypoints(gray_image, keypoints, image)

# Hin th hnh nh
plt.imshow(img_with_keypoints)
plt.show()

# In s lng key points c tm thy
print(f"S lng key points: {len(keypoints)}")
```

### Gii thch c php v cc tham s
- `cv2.SIFT_create()`: Khi to mt i tng SIFT. Bn c th truyn cc tham s nh `nfeatures`  ch nh s lng c trng ti a c tr v.
- `sift.detectAndCompute(image, mask)`: Phng thc ny tm cc key points v tnh ton descriptors cho hnh nh. Tham s `image` l hnh nh u vo v `mask` l mt mt n  ch nh khu vc ca hnh nh  pht hin (None nu khng s dng mt n).
- `cv2.drawKeypoints(image, keypoints, outImage)`: Hm ny v cc key points  tm c ln hnh nh. `image` l hnh nh u vo, `keypoints` l danh sch cc key points, v `outImage` l hnh nh kt qu s hin th cc key points.

### Thay i gi tr tham s
Bn c th thay i cc tham s ca `cv2.SIFT_create()`  xem s thay i v s lng v cht lng cc key points c tm thy. V d, gim `nfeatures` xung cn 50  gii hn s lng c trng ti a c pht hin:
```python
sift = cv2.SIFT_create(nfeatures=50)
```

Hy chc chn rng ng dn n hnh nh l ng v hnh nh  hp l  code c th chy m khng gp li.

Trong OpenCV, c nhiu gii thut v hm h tr cho vic so khp c trng (feature matching) gia cc hnh nh. Hai phng php ph bin l s dng Brute-Force Matcher v FLANN based Matcher. Ti s gii thiu v c hai phng php ny v cch s dng chng  so khp cc c trng c trch xut bi SIFT.

### 1. Brute-Force Matcher
Brute-Force Matcher l phng php n gin nht  so khp c trng. N so snh mi descriptor trong mt tp hp vi mi descriptor trong tp hp khc  tm ra cc cp c trng tt nht.

#### Cc hm h tr:
- `cv2.BFMatcher(normType, crossCheck)`: To mt Brute-Force matcher. `normType` ch nh khong cch  so snh (thng l `cv2.NORM_L2` cho SIFT v SURF, `cv2.NORM_HAMMING` cho ORB v BRIEF). `crossCheck=True` ch nh rng ch nhng cp c trng m mi bn l tt nht ca bn kia mi c tr v.

#### V d minh ha:
```python
import cv2
import matplotlib.pyplot as plt

# c hai hnh nh
img1 = cv2.imread('image1.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)

# Khi to SIFT v tm key points v descriptors
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)

# Khi to Brute-Force matcher v thc hin so khp
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bf.match(des1, des2)

# Sp xp cc matches theo khong cch
matches = sorted(matches, key=lambda x: x.distance)

# V cc kt qu so khp
img_matches = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], None, flags=2)

# Hin th hnh nh
plt.imshow(img_matches)
plt.show()
```

### 2. FLANN Based Matcher
FLANN (Fast Library for Approximate Nearest Neighbors) l mt phng php ti u ha vic tm kim gn nht, nhanh hn Brute-Force khi lm vic vi lng ln d liu.

#### Cc hm h tr:
- `cv2.FlannBasedMatcher(index_params, search_params)`: To mt FLANN matcher. `index_params` v `search_values` l cc tham s cu hnh cho thut ton tm kim.

#### V d minh ha:
```python
import numpy as np

# Cu hnh cho SIFT v FLANN
index_params = dict(algorithm=1, trees=5)
search_params = dict(checks=50)

# Khi to FLANN matcher v thc hin so khp
flann = cv2.FlannBasedMatcher(indexParams, search_params)
matches = flann.knnMatch(des1, des2, k=2)

# p dng Low's ratio test
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

# V cc kt qu so khp
img_matches = cv2.drawMatchesKnn(img1, kp1, img2, kp2, [good_matches], None, flags=2)

# Hin th hnh nh
plt.imshow(img_matches)
plt.show()
```

Trong v d trn, Low's ratio test c s dng  loi b cc matches khng tt, gip ci thin

  chnh xc ca kt qu so khp. C hai v d u minh ha cch so khp c trng gia hai hnh nh s dng SIFT.

Thut ton to nh ton cnh trong OpenCV s dng mt lot cc bc bao gm pht hin c trng, so khp c trng, v cui cng l ghp cc nh li  to thnh nh ton cnh. OpenCV cung cp mt lp tin ch c tn l `Stitcher`  t ng ha qu trnh ny.

### Thut ton To nh Ton Cnh:
1. **Pht hin c trng**: u tin, thut ton pht hin cc c trng trong mi nh. Cc c trng ny l cc im ng ch  c th c so snh mt cch ng tin cy gia cc nh khc nhau.
2. **So Khp c trng**: Sau khi pht hin c trng, thut ton tm kim cc c trng tng ng gia cc cp nh. iu ny cho php thut ton xc nh cc mi lin kt gia cc nh.
3. **c tnh Bin i (Homography)**: Da trn cc c trng  c ghp ni, thut ton c tnh mt m hnh bin i  cn chnh v bin i cc nh sao cho chng ghp kht vi nhau.
4. **Ghp nh**: Cc nh c bin i sau  c ghp li  to thnh nh ton cnh cui cng.

### C Php Cc Hm c S Dng:
OpenCV cung cp lp `Stitcher`  to nh ton cnh. C php c bn:

```python
cv2.Stitcher_create(mode=cv2.Stitcher_PANORAMA)
```
- `mode`: Ch  ca Stitcher, thng l `cv2.Stitcher_PANORAMA` (dnh cho nh ngang) hoc `cv2.Stitcher_SCANS` (dnh cho nh dc).

#### V d Minh Ha:
Di y l on code Python mu  to nh ton cnh t hai hoc nhiu nh.

```python
import cv2
import numpy as np

def create_panorama(images, show_matches=False):
    # Khi to i tng Stitcher
    stitcher = cv2.Stitcher_create(cv2.Stitcher_PANORAMA)
    
    # Ghp cc nh v ly kt qu
    (status, panorama) = stitcher.stitch(images)
    
    if status == cv2.Stitcher_OK:
        # Hin th nh ton cnh
        cv2.imshow("Panorama", panorama)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print("Khng th to nh ton cnh, li:", status)

# c cc nh
img1 = cv2.imread('path_to_image1.jpg')
img2 = cv2.imread('path_to_image2.jpg')

# To nh ton cnh
create_panorama([img1, img2])
```

Trong v d trn, cc nh c c v a vo hm `create_panorama`, s dng i tng `Stitcher`  ghp chng li. Nu qu trnh thnh cng, nh ton cnh s c hin th. Nu khng, chng trnh s in ra m li. m bo cc ng dn n nh l chnh xc v cc nh phi c chiu cao tng t nhau  kt qu ghp c cht lng tt.

````python
import cv2
import matplotlib.pyplot as plt

# c hnh nh
image = cv2.imread('panorama3.jpg')
# image = cv2.imread('panorama4.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Khi to SIFT detector
sift = cv2.SIFT_create()

# Tm key points v descriptors vi SIFT
keypoints, descriptors = sift.detectAndCompute(gray_image, None)

# V key points ln hnh nh
img_with_keypoints = cv2.drawKeypoints(gray_image, keypoints, image)

# Hin th hnh nh
plt.imshow(img_with_keypoints)
plt.show()

# In s lng key points c tm thy
print(f"S lng key points: {len(keypoints)}")
````

````python
import cv2
import matplotlib.pyplot as plt

# c hai hnh nh
img1 = cv2.imread('Panorama1.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('Panorama2.jpg', cv2.IMREAD_GRAYSCALE)

# Khi to SIFT v tm key points v descriptors
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)

# Khi to Brute-Force matcher v thc hin so khp
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bf.match(des1, des2)

# Sp xp cc matches theo khong cch
matches = sorted(matches, key=lambda x: x.distance)

# V cc kt qu so khp
img_matches = cv2.drawMatches(img1, kp1, img2, kp2, matches[:50], None, flags=2)

# Hin th hnh nh
plt.imshow(img_matches)
plt.show()
````

````python
import numpy as np

# Cu hnh cho SIFT v FLANN
index_params = dict(algorithm=1, trees=5)  # S dng thut ton k-d tree
search_params = dict(checks=50)  # S lng vng lp thut ton, gi tr cng ln th kt qu cng chnh xc nhng tn thi gian hn

# Khi to FLANN matcher v thc hin so khp
flann = cv2.FlannBasedMatcher(index_params, search_params)
matches = flann.knnMatch(des1, des2, k=2)

# p dng Low's ratio test
good_matches = []
for m, n in matches:
    if m.distance < 0.75 * n.distance:
        good_matches.append(m)

# V cc kt qu so khp
img_matches = cv2.drawMatchesKnn(img1, kp1, img2, kp2, [good_matches], None, flags=2)

# Hin th hnh nh
plt.imshow(img_matches)
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh
img1 = cv2.imread('panorama4.jpg')
img2 = cv2.imread('panorama5.jpg')
img3 = cv2.imread('panorama3.jpg')

# Danh sch nh v kim tra ti thnh cng
images = [img1, img2, img3]
all_images_loaded = True  # Gi s tt c nh c ti thnh cng

for i, img in enumerate(images, start=1):
    if img is None:
        print(f"Khng th ti nh s {i}")
        all_images_loaded = False
    else:
        # Chuyn i sang RGB  hin th bng Matplotlib
        images[i-1] = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(images[i-1])
        plt.title(f'nh s {i} (RGB)')
        plt.show()

# Ch tip tc nu tt c nh  c ti thnh cng
if all_images_loaded:
    # Khi to Stitcher
    stitcher = cv2.Stitcher_create()

    # Thc hin ghp nh
    ret, pano = stitcher.stitch(images)

    if ret == cv2.Stitcher_OK:
        # Hin th nh ton cnh
        plt.imshow(pano)
        plt.title('nh ton cnh')
        plt.show()
    else:
        print("Ghp nh tht bi:", ret)
else:
    print("Mt hoc nhiu nh khng c ti ln thnh cng, kim tra li ng dn hoc tn file.")
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# c nh
img1 = cv2.imread('panorama6.png')
img2 = cv2.imread('panorama7.png')
img3 = cv2.imread('panorama8.png')

# Danh sch nh v kim tra ti thnh cng
images = [img1, img2, img3]
all_images_loaded = True  # Gi s tt c nh c ti thnh cng

# S dng SIFT  tm im c trng
sift = cv2.SIFT_create()
keypoints = []
descriptors = []

for i, img in enumerate(images, start=1):
    if img is None:
        print(f"Khng th ti nh s {i}")
        all_images_loaded = False
    else:
        kp, des = sift.detectAndCompute(img, None)
        keypoints.append(kp)
        descriptors.append(des)
        img_with_kp = cv2.drawKeypoints(img, kp, None)
        plt.imshow(img_with_kp)
        plt.title(f'nh s {i} vi im c trng')
        plt.show()
        print(f"S im c trng trong nh s {i}: {len(kp)}")

# Ch tip tc nu tt c nh  c ti thnh cng
if all_images_loaded:
    # S dng BFMatcher  so khp im c trng gia cc cp nh
    bf = cv2.BFMatcher()
    for i in range(len(images) - 1):
        matches = bf.knnMatch(descriptors[i], descriptors[i+1], k=2)
        # p dng Low's ratio test
        good = []
        for m, n in matches:
            if m.distance < 0.75 * n.distance:
                good.append(m)
        print(f"S matches tt gia nh {i+1} v {i+2}: {len(good)}")

    # Khi to Stitcher vi cu hnh ty chnh
    stitcher = cv2.Stitcher_create()
    stitcher.setPanoConfidenceThresh(0.6)  # iu chnh ngng tin cy

    # Thc hin ghp nh
    ret, pano = stitcher.stitch(images)

    if ret == cv2.Stitcher_OK:
        # Hin th nh ton cnh
        plt.imshow(pano)
        plt.title('nh ton cnh')
        plt.show()
    else:
        print("Ghp nh tht bi, m li:", ret)
else:
    print("Mt hoc nhiu nh khng c ti ln thnh cng, kim tra li ng dn hoc tn file.")
````

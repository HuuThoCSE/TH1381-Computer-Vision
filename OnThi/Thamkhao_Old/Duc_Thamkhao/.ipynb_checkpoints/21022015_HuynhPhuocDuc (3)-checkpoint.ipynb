```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

1.Tìm hiểu và trình bày những hiểu biết của bạn về biến đổi Hough (Hough Transform);
Trình bày cú pháp hàm tương ứng trong OpenCV cho bài toán phát hiện đường thẳng 
dựa trên biến đổi Hough; Cho ví dụ minh họa, giải thích ý nghĩa các câu lệnh được sử
dụng; Kiểm tra đoạn chương trình minh họa với nhiều hình ảnh khác nhau.

-Biến đổi Hough là phương pháp trong xử lý ảnh để phát hiện bất kì hình dạng nào mà có thể biểu diễn dưới dạng toán học, kể cả nó có bị vở hay bóp méo
-Cách hoạt động:
    +Tạo mảng 2 chiều và để toàn bộ giá trị bằng 0
    +Đặt hàng là giá trị r và cột là giá trị theta-(giá trị góc)
    +Kích thước của mảng phù thuộc độ chính xác bạn muốn. Bạn muốn độ chính xác của giá trị góc lên đến từng độ, bạn cần tới 180 cột(góc tối đa của 1 đường thẳng là 180)
    +Đối tham số r, kích thức tối đa có thể đạt được là chiều dài đường chéo của hình ảnh.Vậy nên chỉ cần 1 pixel chính xác, số lượng dòng dữ liệu có thể là độ dài đường chéo của ảnh đó


Cú pháp:
lines = cv2.HoughLines(edges, 1, np.pi/180, 200)
    edge: Là ảnh đầu vào, nên là ảnh nhị phân để có áp dụng ngưỡng phát hiện cạnh, trước khi áp dụng biến đổi Hough để tìm kiếm
    Tham số thứ hai(1)và ba(np.pi/180): theo thứ tự lần lượt là giá trị r(khoảng cách) và theta(giá trị góc)
    Tham số thứ tư(200): là giá trị ngưỡng, có tối thiểu giá trị đề xuất được xác định là ngưỡng
Số lượng của các đề xuất phù thuộc vào số điểm trên đường thẳng. Vì vậy nó chỉ biểu diễn một lượng tối thiểu chiều dài của đường thẳng  được phát hiện

```python


# Tải ảnh lên
img = cv2.imread('sudoku.jpg')

# Chuyển đổi sang ảnh xám
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Phát hiện cạnh với canny
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Lấy mảng giá trị của r và theta
lines = cv2.HoughLines(edges, 1, np.pi/180, 200)

# Duyệt qua tất cả giá trị trong mảng trên
for r_theta in lines:
	arr = np.array(r_theta[0], dtype=np.float64)
	r, theta = arr
	# lưu giá trị cos(theta) in a
	a = np.cos(theta)

	# lưu giá trị sin(theta) in b
	b = np.sin(theta)

	#lưu giá trị rcos(theta) vào x0 
	x0 = a*r

	#lưu giá trị rsin(theta) vào y0 
	y0 = b*r

	# x1 là giá trị làm tròn (rcos(theta)-1000sin(theta))
	x1 = int(x0 + 1000*(-b))

	# y1 là giá trị làm tròn (rsin(theta)+1000cos(theta))
	y1 = int(y0 + 1000*(a))

	# x2 là giá trị làm tròn (rcos(theta)+1000sin(theta))
	x2 = int(x0 - 1000*(-b))

	# y2 là giá trị làm tròn (rsin(theta)-1000cos(theta))
	y2 = int(y0 - 1000*(a))

	# vẽ các đường thẳng dựa trên các điểm được xác định phía trên
	cv2.line(img, (x1, y1), (x2, y2), (255, 0,0), 2)

# All the changes made in the input image are finally
# written on a new image houghlines.jpg
cv2.imwrite('linesDetected_sudoku.jpg', img)
plt.imshow(img, cmap='gray')
```

2. Trình bày cú pháp hàm tương ứng trong OpenCV cho phát hiện các đường tròn trong 
ảnh; Cho ví dụ minh họa, giải thích ý nghĩa các câu lệnh được sử dụng; Kiểm tra đoạn 
chương trình minh họa với nhiều hình ảnh khác nhau.

detected_circles = cv2.HoughCircles(gray_blurred, cv2.HOUGH_GRADIENT, 1, 20, param1 = 50, param2 = 30, minRadius = 1, maxRadius = 40)
    +gray_blurred: ảnh đầu vào (ảnh xám và được làm mịn đi)
    +cv2.HOUGH_GRADIENT: phương pháp phát hiện bằng cạnh bằng gradient
    + 1 : là dp tham số tỉ lệ độ phân giải của ảnh gốc được
    + 20 : là minDist là tham số điều khiển khoản cách tối thiểu  giữa các hình tròn phát hiện
    + param1: dùng để set ngưỡng trên của hàm phát hiện cạnh canny còn ngưỡng dưới sẽ là tham số đó chia 2
    + param2: là ngưỡng tích lũy cho các hình tròn đề xuất. Bằng cách tăng giá trị này, có thể chắc chắn là kết quả chỉ có nhũng hình tròn hoàn thiện nhất
    + minRadius: bán kính tối thiểu của hình tròn
    + maxRadius: Bán kính tối đa của hình tròn

```python


img = cv2.imread('puppy1.jpg', cv2.IMREAD_COLOR) 
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

# Làm mịn ảnh sủ dụng 3 * 3 kernel. 
gray_blurred = cv2.blur(gray, (3, 3)) 

# Sử dụng biến đổi Hough để phát hiện hình tròn
detected_circles = cv2.HoughCircles(gray_blurred, 
				cv2.HOUGH_GRADIENT, 1, 20, param1 = 50, 
			param2 = 50, minRadius = 1, maxRadius = 40) 

# Vẽ các hình tròn được phát hiện. 
if detected_circles is not None: 

	# Chuyển đổi các tham số của hình tròn a, b và r sang kiểu số nguyên. 
	detected_circles = np.uint16(np.around(detected_circles)) 

	for pt in detected_circles[0, :]: 
		a, b, r = pt[0], pt[1], pt[2] 

		# Vẽ hình tròn. 
		cv2.circle(img, (a, b), r, (0, 255, 0), 2) 

		# Vẽ 1 chấm tròn để hiển thị tâm. 
		cv2.circle(img, (a, b), 1, (0, 0, 255), 3)

cv2.imwrite('detect_circle.jpg',img) 
plt.imshow(img, cmap='gray') 
```

1.Tìm hiểu về giải thuật phát hiện góc Harris; Trình bày cú pháp hàm tương ứng trong 
OpenCV cho bài toán phát hiện góc bằng giải thuật Harris; Cho ví dụ minh họa, giải 
thích ý nghĩa các câu lệnh được sử dụng

Là thuật toán được phát triển để xác định các góc bên trong hình ảnh. Các góc trong ảnh cơ bản xác định như vùng mà trong đó có sự thay đổi dột ngột về cường độ gradient ở các chiều các hướng. Góc được trích xuất có thể là thành phần của đặc trưng ảnh, có thể so khớp với đặc trưng cảu ảnh khác và có thể dung để trích ra những thông tin chính xác từ ảnh đầu vào

Cú pháp:
cv2.cornerHarris(src, dest, blockSize, kSize, freeParameter, borderType)
    +src: ảnh đầu vào
    +dest: ảnh lưu bởi đầu ra của hàm phát hiện Harris, kích thước giống với ảnh đầu vào
    +blockSize: kích thước liền kề(ở mỗi pixel khoản cách liền kề (blockSize*blockSize)được xác định)
    +ksize: Tham số cho Sobel()
    +freeParameter: Tham số tự do của Harris
    +borderType : Phương pháp ngoại suy-trả về vị trí của pixel tương ứng

```python


image = cv2.imread('sudoku3.jpg') 
image= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

operatedImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 


operatedImage = np.float32(operatedImage) 

# Sử dụng giải pháp Harris để phát hiện góc
dest = cv2.cornerHarris(operatedImage, 2, 5, 0.07) 

# Kết quả được thông qua duỗi góc 
dest = cv2.dilate(dest, None) 

# Đổi ngược về ảnh gốc , với giá trị ngưỡng được tối ưu.
image[dest > 0.01 * dest.max()]=[255, 0, 0] 

# the window showing output image with corners 
cv2.imwrite('detect_corner.jpg', image) 
plt.imshow(image, cmap='gray')

# De-allocate any associated memory usage 

```

2.Tìm hiểu về giải thuật phát hiện góc Shi-Tomasi; Trình bày cú pháp hàm tương ứng 
trong OpenCV cho bài toán phát hiện góc bằng giải thuật Shi-Tomasi; Cho ví dụ minh 
họa, giải thích ý nghĩa các câu lệnh được sử dụng.

Nói các góc có thể được phát hiện bằng cách tìm kiếm những thay đổi đáng kể ở mọi hưởng.
Xem như ta có 1 cửa sổ vuông nhỏ trên hình để có quét qua toàn bộ ảnh tìm góc
Trượt cửa sổ vuông nhỏ đó ở bất kì hướng nào sẽ cho ra sự thay đổi lớn về hình dáng-> Phát hiện dược góc


Cú pháp:
cv2.goodFeaturesToTrack(gray_img, maxc, Q, minD)
    +gray_img: ảnh đầu vào là dạng ảnh xám
    +Maxc: số lượng góc cạnh tối đa bạn muốn phát hiện(cho số âm sẽ lấy tất cả)
    +Q: Mức độ chất lượng tham số (mặc định: 0.01)
    +maxD: Khoản cách tối đa (mặc định: 10)

```python

img = cv2.imread('sudoku3.jpg') 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

# sử dụng phát hiện góc Shi-Tomasi 
corners = cv2.goodFeaturesToTrack(gray_img, 300, 0.01, 2) 

# Chuyển đổi những góc này thành giá trị nguyên- để vẽ lên ảnh gốc 
corners = np.int0(corners) 

# draw red color circles on all corners 
for i in corners: 
	x, y = i.ravel() 
	cv2.circle(img, (x, y), 4, (255, 0, 0), -1) 

cv2.imwrite('detect_corner.jpg', img)
# resulting image 
plt.imshow(img) 

```

3. So sánh kết quả giữa 2 giải thuật trên cho bài toán phát hiện góc

Sau khi xem xong kết quả em thấy 
Harris:
    Số lượng góc phát hiện nhiều hơn
    Phát hiện góc ở cả chi nhỏ
    ít bỏ sót
    có phát hiện những phần không phải là góc
    mật độ có thể rất dày ở một số trường hợp
    các tham số phức tạp
 Shi-Tomasi:
    Số lượng góc phát hiện ít hơn
    không nhận dạng được ở vị trí mờ
    Dễ bỏ qua các góc ảnh
    khó nhận ra ở các chi tiết nhỏ
    Các tham số dễ hiểu dễ tác động vào cho người dùng

Việc xác định các góc góp phần hỗ trợ giúp xác định rõ vật thể trong ảnh bì mờ, có độ phân giải không cao

1.Tìm hiểu và trình bày lý thuyết về đường bao trong ảnh; Trình bày và giải thích cú 
pháp các hàm hỗ trợ trích xuất và vẽ đường bao trong OpenCV; Cho ví dụ minh họa

Cơ bản Đường biên là đường nối các điểm ngoại vi của vật thể. 

	cv.findContours(image, mode, method[, contours[, hierarchy[, offset]]])
    image: ảnh đầu vào
    mode: chế độ thu hồi biên
    method: phương thức xấp xỉ đường biên
        coutour: Phát hiện biên mỗi contour được lưu trữ ở mỗi vector
        hierarchy: chọn đầu ra cho vector
        offset: chọn offset


```python
im = cv2.imread('sudoku.jpg')
assert im is not None, "file could not be read, check with os.path.exists()"
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(im, contours, -1, (0, 255, 0), 3)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
plt.show()
```

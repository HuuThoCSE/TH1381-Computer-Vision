Vit hm gradientEdges cho php lc tch bin nh theo phng php Gradient vi 
b lc ty chn Robert, Prewitt, Sobel

```python
def gradientEdges(anh, boloc):
    title=["nh gc", "nh ton bin", "nh bin dc", "nh bin ngang"]
    if(boloc=="Sobel"):
        # to 2 mc n x y sobel
        kernal_x=np.array([[-1, 0, 1],
                             [-2, 0, 2],
                             [-1, 0, 1]], dtype=np.float32)
        kernal_y=np.array([[-1, -2, -1],
                             [0, 0, 0],
                             [1, 2, 1]], dtype=np.float32)
        # lc nh ly bin ngang v mc n ly bin dc 
        anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
        anh_fll_y = cv2.filter2D(anh, -1, kernal_y)
        # cng li  ly nh ton bin
        anh_fll_xy=anh_fll_x + anh_fll_y
        listsobel=[anh,anh_fll_xy,anh_fll_y, anh_fll_x]
        for i in range(len(listsobel)):
            plt.subplot(2,2,i+1) 
            plt.imshow(listsobel[i],cmap = 'gray') 
            plt.title(title[i]) 
    elif(boloc=="Prewitt"):
        # to 2 mc n x y prewitt
        kernal_x=np.array([[-1, 0, 1],
                             [-1, 0, 1],
                             [-1, 0, 1]], dtype=np.float32)
        kernal_y=np.array([[-1, -1, -1],
                             [0, 0, 0],
                             [1, 1, 1]], dtype=np.float32)
        # lc nh ly bin ngang v mc n ly bin dc 
        anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
        anh_fll_y = cv2.filter2D(anh, -1, kernal_y)
        # cng li  ly nh ton bin
        anh_fll_xy=anh_fll_x + anh_fll_y
        listprewitt=[anh,anh_fll_xy,anh_fll_y, anh_fll_x]
        for i in range(len(listprewitt)):
            plt.subplot(2,2,i+1) 
            plt.imshow(listprewitt[i],cmap = 'gray')
            plt.title(title[i]) 
    elif(boloc=="Robert"):
        # to 2 mc n x y robert
        kernal_x=np.array([[1, 0],
                             [0, -1]], dtype=np.float32)
        kernal_y=np.array([[0, 1],
                             [-1, 0]], dtype=np.float32)
        # lc nh ly bin ngang v mc n ly bin dc 
        anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
        anh_fll_y = cv2.filter2D(anh, -1, kernal_y)
        # cng li  ly nh ton bin
        anh_fll_xy=anh_fll_x + anh_fll_y
        listrobert=[anh,anh_fll_xy,anh_fll_y, anh_fll_x]
        for i in range(len(listrobert)):
            plt.subplot(2,2,i+1) 
            plt.imshow(listrobert[i],cmap = 'gray')
            plt.title(title[i]) 
```

```python


def canny_edge_detection(frame): 
	# Chuyn cc khung hnh sang nh xm  s dng trong pht hin bin 
	
	
	# p dng thut ton lm m Gaussian -> gim nhiu lm mt bin 
	blurred = cv2.GaussianBlur(src=frame, ksize=(3, 5), sigmaX=0.5) 
	
	# dng Canny  pht hin bin 
	edges = cv2.Canny(blurred, 70, 135) 
	# tr v nh  lm m v nh pht hin cnh
	return blurred, edges
```

```python
def cannyedge(anh):    
    blurred = cv2.GaussianBlur(src=anh, ksize=(3, 5), sigmaX=0.5)
    kernal_x=np.array([[-1, 0, 1],
                             [-2, 0, 2],
                             [-1, 0, 1]], dtype=np.float32)
    kernal_y=np.array([[-1, -2, -1],
                             [0, 0, 0],
                             [1, 2, 1]], dtype=np.float32)
    # lc nh ly bin ngang v mc n ly bin dc 
    anh_fll_x = cv2.filter2D(blurred, -1, kernal_x)
    anh_fll_y = cv2.filter2D(blurred, -1, kernal_y)
    # cng li  ly nh ton bin
    anh_fll_xy=anh_fll_x + anh_fll_y 
    max_v=np.max(anh_fll_xy)
    # ln gradient
    gradient_magnitude = np.sqrt(anh_fll_x**2 + anh_fll_y**2)
    #Hng gradient
    gradient_orientation = np.arctan2(anh_fll_y, anh_fll_x)
    #to nh en vi kch thc ca nh sobel
    suppressed_img = np.zeros_like(anh_fll_xy)
    rows, cols = anh_fll_xy.shape
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            #ly  ln v hng ti tng im
            angle = gradient_orientation[i, j]
            mag = gradient_magnitude[i, j]

            # Tm pixels gn k theo hng gradient
            if (-np.pi / 8 <= angle < np.pi / 8) or (7 * np.pi / 8 <= angle <= np.pi) or (-7 * np.pi / 8 <= angle < -np.pi / 8):
                neighbor1 = gradient_magnitude[i, j + 1]
                neighbor2 = gradient_magnitude[i, j - 1]
            elif (np.pi / 8 <= angle < 3 * np.pi / 8) or (-7 * np.pi / 8 <= angle < -5 * np.pi / 8):
                neighbor1 = gradient_magnitude[i - 1, j + 1]
                neighbor2 = gradient_magnitude[i + 1, j - 1]
            elif (3 * np.pi / 8 <= angle < 5 * np.pi / 8) or (-5 * np.pi / 8 <= angle < -3 * np.pi / 8):
                neighbor1 = gradient_magnitude[i - 1, j]
                neighbor2 = gradient_magnitude[i + 1, j]
            else:
                neighbor1 = gradient_magnitude[i - 1, j - 1]
                neighbor2 = gradient_magnitude[i + 1, j + 1]

            # Loi b pixels gn k nu khng t   ln ca pixel ang xt
            if mag <= neighbor1 and mag <= neighbor2:
                suppressed_img[i, j] = mag
            elif mag > neighbor1 and mag > neighbor2:
                suppressed_img[i, j] = 0

    min_h=70
    max_h=135
    for row in suppressed_img:
        for pixel in row:
            if pixel > max_h:
                pixel = 255
            if max_h<pixel > min_h:
                pixel = pixel
            if pixel < min_h:
                pixel = 0

    title=["nh gc", "nh bin"]
    listlap=[anh,suppressed_img]
    for i in range(len(listlap)):
        plt.subplot(1,2,i+1) 
        plt.imshow(listlap[i],cmap = 'gray') 
        plt.title(title[i])
    
    
```

```python
def open_a(anh):
    _,binary_img = cv2.threshold(anh, 127, 255, cv2.THRESH_BINARY)
    kernel = np.ones((5, 5), np.uint8) 
    img_erosion = cv2.erode(binary_img, kernel, iterations=1) 
    img_dilation = cv2.dilate(img_erosion, kernel, iterations=1) 
    return img_dilation

def close_a(anh):
    _, binary_img = cv2.threshold(anh, 127, 255, cv2.THRESH_BINARY)
    kernel = np.ones((5, 5), np.uint8) 
    img_dilation = cv2.dilate(binary_img, kernel, iterations=1) 
    img_erosion = cv2.erode(img_dilation, kernel, iterations=1) 
    
    return img_erosion
```

Nghin cu cc link sau v trnh by cch thc openCV h tr m/ng nh c mask.

H tr M:
    hm h tr
    cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
        image : Mng hnh nh u vo. 
        cv2.MORPH_OPEN : p dng thao tc M hnh thi. 
        kernel : Phn t cu trc.

 

```python
# Python program to illustrate 
# Opening morphological operation 
# on an image 

# organizing imports 
import cv2 
import numpy as np 

# return video from the first webcam on your computer. 
screenRead = cv2.VideoCapture(0) 

# loop runs if capturing has been initialized. 
while(1): 
	# reads frames from a camera 
	_, image = screenRead.read() 
	
	# Converts to HSV color space, OCV reads colors as BGR 
	# frame is converted to hsv 
	hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
	
	# defining the range of masking 
	blue1 = np.array([110, 50, 50]) 
	blue2 = np.array([130, 255, 255]) 
	
	# initializing the mask to be 
	# convoluted over input image 
	mask = cv2.inRange(hsv, blue1, blue2) 

	# passing the bitwise_and over 
	# each pixel convoluted 
	res = cv2.bitwise_and(image, image, mask = mask) 
	
	# defining the kernel i.e. Structuring element 
	kernel = np.ones((5, 5), np.uint8) 
	
	# defining the opening function 
	# over the image and structuring element 
	opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel) 
	
	# The mask and opening operation 
	# is shown in the window 
	cv2.imshow('Mask', mask) 
	cv2.imshow('Opening', opening) 
	
	# Wait for 'a' key to stop the program 
	if cv2.waitKey(1) & 0xFF == ord('a'): 
		break

# De-allocate any associated memory usage 
cv2.destroyAllWindows() 

# Close the window / Release webcam 
screenRead.release() 
```

Viết hàm gradientEdges cho phép lọc tách biên ảnh theo phương pháp Gradient với 
bộ lọc tùy chọn Robert, Prewitt, Sobel

```python
def gradientEdges(anh, boloc):
    title=["Ảnh gốc", "Ảnh toàn biên", "Ảnh biên dọc", "Ảnh biên ngang"]
    if(boloc=="Sobel"):
        # tạo 2 mặc nạ x y sobel
        kernal_x=np.array([[-1, 0, 1],
                             [-2, 0, 2],
                             [-1, 0, 1]], dtype=np.float32)
        kernal_y=np.array([[-1, -2, -1],
                             [0, 0, 0],
                             [1, 2, 1]], dtype=np.float32)
        # lọc ảnh lấy biên ngang và mặc nạ lấy biên dọc 
        anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
        anh_fll_y = cv2.filter2D(anh, -1, kernal_y)
        # cộng lại để lấy ảnh toàn biên
        anh_fll_xy=anh_fll_x + anh_fll_y
        listsobel=[anh,anh_fll_xy,anh_fll_y, anh_fll_x]
        for i in range(len(listsobel)):
            plt.subplot(2,2,i+1) 
            plt.imshow(listsobel[i],cmap = 'gray') 
            plt.title(title[i]) 
    elif(boloc=="Prewitt"):
        # tạo 2 mặc nạ x y prewitt
        kernal_x=np.array([[-1, 0, 1],
                             [-1, 0, 1],
                             [-1, 0, 1]], dtype=np.float32)
        kernal_y=np.array([[-1, -1, -1],
                             [0, 0, 0],
                             [1, 1, 1]], dtype=np.float32)
        # lọc ảnh lấy biên ngang và mặc nạ lấy biên dọc 
        anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
        anh_fll_y = cv2.filter2D(anh, -1, kernal_y)
        # cộng lại để lấy ảnh toàn biên
        anh_fll_xy=anh_fll_x + anh_fll_y
        listprewitt=[anh,anh_fll_xy,anh_fll_y, anh_fll_x]
        for i in range(len(listprewitt)):
            plt.subplot(2,2,i+1) 
            plt.imshow(listprewitt[i],cmap = 'gray')
            plt.title(title[i]) 
    elif(boloc=="Robert"):
        # tạo 2 mặc nạ x y robert
        kernal_x=np.array([[1, 0],
                             [0, -1]], dtype=np.float32)
        kernal_y=np.array([[0, 1],
                             [-1, 0]], dtype=np.float32)
        # lọc ảnh lấy biên ngang và mặc nạ lấy biên dọc 
        anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
        anh_fll_y = cv2.filter2D(anh, -1, kernal_y)
        # cộng lại để lấy ảnh toàn biên
        anh_fll_xy=anh_fll_x + anh_fll_y
        listrobert=[anh,anh_fll_xy,anh_fll_y, anh_fll_x]
        for i in range(len(listrobert)):
            plt.subplot(2,2,i+1) 
            plt.imshow(listrobert[i],cmap = 'gray')
            plt.title(title[i]) 
```

với bộ lọc tùy chọn là 1 trong 4 bộ lọc Laplacian đã giới thiệu (theo đúng thứ tự). 

```python
def laplaciantEdges(anh, boloc):
   title=["Ảnh gốc", "Ảnh biên"] 
   if(boloc==1):
      kernal_x=np.array([[0, 1, 0],
                             [1, -4, 1],
                             [0, 1, 0]], dtype=np.float32)
      anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
      listlap=[anh,anh_fll_x]
      for i in range(len(listlap)):
         plt.subplot(1,2,i+1) 
         plt.imshow(listlap[i],cmap = 'gray') 
         plt.title(title[i])
   elif(boloc==2):
      kernal_x=np.array([[1, 1, 1],
                             [1, -8, 1],
                             [1, 1, 1]], dtype=np.float32)
      anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
      listlap=[anh,anh_fll_x]
      for i in range(len(listlap)):
         plt.subplot(1,2,i+1) 
         plt.imshow(listlap[i],cmap = 'gray') 
         plt.title(title[i])
   elif(boloc==3):
      kernal_x=np.array([[0, -1, 0],
                             [-1, 4, -1],
                             [0, -1, 0]], dtype=np.float32)
      anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
      listlap=[anh,anh_fll_x]
      for i in range(len(listlap)):
         plt.subplot(1,2,i+1) 
         plt.imshow(listlap[i],cmap = 'gray') 
         plt.title(title[i])   
   elif(boloc==4):
      kernal_x=np.array([[-1,-1, -1],
                             [-1, 8, -1],
                             [-1, -1, -1]], dtype=np.float32)
      anh_fll_x = cv2.filter2D(anh, -1, kernal_x)
      listlap=[anh,anh_fll_x]
      for i in range(len(listlap)):
         plt.subplot(1,2,i+1) 
         plt.imshow(listlap[i],cmap = 'gray') 
         plt.title(title[i])
```

Nghiên cứu link và trình bày các bước thực hiện lọc tách biên với bộ lọc tối ưu Canny:
    B1-giảm nhiểu:
    Giảm nhiểu bằng cách làm mờ ảnh, để làm mờ ta dùng bộ lọc Gaussian(lấy trọng số trung bình cho mỗi pixel), giảm độ nhạy do sự thay đổi về cường độ. Mặt nạ càng lớn làm mờ càng hiệu quả
    B2- tìm cường độ gradient:
    Tìm cường độ giúp cho việc xác định các vị trí có sự dịch chuyển về cường độ đột ngột-biểu diễn cho các đường biên của ảnh. Về kĩ thuật ta sẽ dùng thuật toán sobel để xác định các cường độ như thế
    B3- triệt tiêu thừa:
    Dùng để làm cho các cạnh mỏng đi.Mỗi pixel sẽ được xem xét hướng gradient để xác định xem nó có phải một thành phần của đường biên không
    B4- Phân ngưỡng :
    Thiết lập các ngưỡng trên và ngưỡng dưới, cường độ gradient lớn hơn ngưỡng trên-> chắc chắn là cạnh, nhỏ hơn-> không là cạnh và bị loại bỏ, nằm giữa hai ngưỡng-> có thể là cạnh

Viết đoạn chương trình thực hiện lọc tách biên với bộ lọc Canny. Giải thích ý nghĩa 
các câu lệnh được sử dụng

```python


def canny_edge_detection(frame): 
	# Chuyển các khung hình sang ảnh xám để sử dụng trong phát hiện biên 
	
	
	# Áp dụng thuật toán làm mờ Gaussian -> giảm nhiểu làm mượt biên 
	blurred = cv2.GaussianBlur(src=frame, ksize=(3, 5), sigmaX=0.5) 
	
	# dùng Canny để phát hiện biên 
	edges = cv2.Canny(blurred, 70, 135) 
	# trả về ảnh đã làm mờ và ảnh phát hiện cạnh
	return blurred, edges
```

```python
def main(): 
	# Mở webcam 
	cap = cv2.VideoCapture(0) 
	
	while True: 
		# Đọc các khung hình của webcam 
		ret, frame = cap.read() 
		if not ret: 
			print('Image not captured') 
			break
		
		# biểu diễn khả năng phát hiện biên trên khung hình 
		blurred, edges = canny_edge_detection(frame) 
		
		# THienr thị khung gốc và khung phát hiện
		cv2.imshow("Original", frame) 
		# cv2.imshow("Blurred", blurred) 
		cv2.imshow("Edges", edges) 
		
		
		if cv2.waitKey(1) & 0xFF == ord('q'): 
			break
	
	
	cap.release() 
	cv2.destroyAllWindows()
```

```python
def cannyedge(anh):    
    blurred = cv2.GaussianBlur(src=anh, ksize=(3, 5), sigmaX=0.5)
    kernal_x=np.array([[-1, 0, 1],
                             [-2, 0, 2],
                             [-1, 0, 1]], dtype=np.float32)
    kernal_y=np.array([[-1, -2, -1],
                             [0, 0, 0],
                             [1, 2, 1]], dtype=np.float32)
    # lọc ảnh lấy biên ngang và mặc nạ lấy biên dọc 
    anh_fll_x = cv2.filter2D(blurred, -1, kernal_x)
    anh_fll_y = cv2.filter2D(blurred, -1, kernal_y)
    # cộng lại để lấy ảnh toàn biên
    anh_fll_xy=anh_fll_x + anh_fll_y 
    max_v=np.max(anh_fll_xy)
    #độ lớn gradient
    gradient_magnitude = np.sqrt(anh_fll_x**2 + anh_fll_y**2)
    #Hướng gradient
    gradient_orientation = np.arctan2(anh_fll_y, anh_fll_x)
    #tạo ảnh đen với kích thước của ảnh sobel
    suppressed_img = np.zeros_like(anh_fll_xy)
    rows, cols = anh_fll_xy.shape
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            #lấy độ lớn và hướng tại từng điểm
            angle = gradient_orientation[i, j]
            mag = gradient_magnitude[i, j]

            # Tìm pixels gần kề theo hướng gradient
            if (-np.pi / 8 <= angle < np.pi / 8) or (7 * np.pi / 8 <= angle <= np.pi) or (-7 * np.pi / 8 <= angle < -np.pi / 8):
                neighbor1 = gradient_magnitude[i, j + 1]
                neighbor2 = gradient_magnitude[i, j - 1]
            elif (np.pi / 8 <= angle < 3 * np.pi / 8) or (-7 * np.pi / 8 <= angle < -5 * np.pi / 8):
                neighbor1 = gradient_magnitude[i - 1, j + 1]
                neighbor2 = gradient_magnitude[i + 1, j - 1]
            elif (3 * np.pi / 8 <= angle < 5 * np.pi / 8) or (-5 * np.pi / 8 <= angle < -3 * np.pi / 8):
                neighbor1 = gradient_magnitude[i - 1, j]
                neighbor2 = gradient_magnitude[i + 1, j]
            else:
                neighbor1 = gradient_magnitude[i - 1, j - 1]
                neighbor2 = gradient_magnitude[i + 1, j + 1]

            # Loại bỏ pixels gần kề nếu không đạt đủ độ lớn của pixel đang xét
            if mag <= neighbor1 and mag <= neighbor2:
                suppressed_img[i, j] = mag
            elif mag > neighbor1 and mag > neighbor2:
                suppressed_img[i, j] = 0

    min_h=70
    max_h=135
    for row in suppressed_img:
        for pixel in row:
            if pixel > max_h:
                pixel = 255
            if max_h<pixel > min_h:
                pixel = pixel
            if pixel < min_h:
                pixel = 0

    title=["Ảnh gốc", "Ảnh biên"]
    listlap=[anh,suppressed_img]
    for i in range(len(listlap)):
        plt.subplot(1,2,i+1) 
        plt.imshow(listlap[i],cmap = 'gray') 
        plt.title(title[i])
    
    
```

```python
def open_a(anh):
    _,binary_img = cv2.threshold(anh, 127, 255, cv2.THRESH_BINARY)
    kernel = np.ones((5, 5), np.uint8) 
    img_erosion = cv2.erode(binary_img, kernel, iterations=1) 
    img_dilation = cv2.dilate(img_erosion, kernel, iterations=1) 
    return img_dilation

def close_a(anh):
    _, binary_img = cv2.threshold(anh, 127, 255, cv2.THRESH_BINARY)
    kernel = np.ones((5, 5), np.uint8) 
    img_dilation = cv2.dilate(binary_img, kernel, iterations=1) 
    img_erosion = cv2.erode(img_dilation, kernel, iterations=1) 
    
    return img_erosion
```

```python
o_a=open_a(img)
c_a=close_a(img)
title=["Ảnh gốc", "Ảnh mở", "Ảnh đóng"]
listlap=[img,o_a,c_a]
for i in range(len(listlap)):
    plt.subplot(1,3,i+1) 
    plt.imshow(listlap[i],cmap = 'gray') 
    plt.title(title[i])
```

Nghiên cứu các link sau và trình bày cách thức openCV hỗ trợ mở/đóng ảnh có mask.

Hỗ trợ Mở:
    hàm hỗ trợ
    cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
        image : Mảng hình ảnh đầu vào. 
        cv2.MORPH_OPEN : Áp dụng thao tác Mở hình thái. 
        kernel : Phần tử cấu trúc.

 

```python
# Python program to illustrate 
# Opening morphological operation 
# on an image 

# organizing imports 
import cv2 
import numpy as np 

# return video from the first webcam on your computer. 
screenRead = cv2.VideoCapture(0) 

# loop runs if capturing has been initialized. 
while(1): 
	# reads frames from a camera 
	_, image = screenRead.read() 
	
	# Converts to HSV color space, OCV reads colors as BGR 
	# frame is converted to hsv 
	hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
	
	# defining the range of masking 
	blue1 = np.array([110, 50, 50]) 
	blue2 = np.array([130, 255, 255]) 
	
	# initializing the mask to be 
	# convoluted over input image 
	mask = cv2.inRange(hsv, blue1, blue2) 

	# passing the bitwise_and over 
	# each pixel convoluted 
	res = cv2.bitwise_and(image, image, mask = mask) 
	
	# defining the kernel i.e. Structuring element 
	kernel = np.ones((5, 5), np.uint8) 
	
	# defining the opening function 
	# over the image and structuring element 
	opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel) 
	
	# The mask and opening operation 
	# is shown in the window 
	cv2.imshow('Mask', mask) 
	cv2.imshow('Opening', opening) 
	
	# Wait for 'a' key to stop the program 
	if cv2.waitKey(1) & 0xFF == ord('a'): 
		break

# De-allocate any associated memory usage 
cv2.destroyAllWindows() 

# Close the window / Release webcam 
screenRead.release() 
```

```python
from skimage import io
import matplotlib.pyplot as plt
```

```python
from PIL import Image
```

```python
image = io.imread('/content/stinkbug.png')
plt.imshow(image)
plt.show()
```

```python
image = cv2.imread('/content/stinkbug.png')
```

```python
plt.imshow(image)
plt.show()
```

```python
image_convC = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_convC)
```

```python
img = cv2.line(image,(0,0),(511,511),(255,0,0),5)

# Drawing Rectangle
img = cv2.rectangle(image,(384,0),(510,128),(0,255,0),3)

# Drawing Circle
img = cv2.circle(image,(447,63), 63, (0,0,255), -1)

# Drawing Ellipse
img = cv2.ellipse(image,(256,256),(100,50),0,0,180,255,-1)

# Drawing Polygon
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
img = cv2.polylines(image,[pts],True,(0,255,255))

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'OpenCV',(10,500), font, 10,(0,0,0),2,cv2.LINE_AA);

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
#scaling
img_resc2 = cv2.resize(image,None,fx=1.5, fy=1.5, interpolation = cv2.INTER_CUBIC)

cv2_imshow(image)

cv2_imshow(img_resc2)
```

```python
#rotation
rows,cols,ch = image.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),120,1)
rota_img = cv2.warpAffine(image,M,(cols,rows))
cv2_imshow(image)
cv2_imshow(rota_img)
```

```python
#shifting
N = np.float32([[1,0,100],[0,1,50]])
shift_img = cv2.warpAffine(image,N,(cols,rows))
cv2_imshow(image)
cv2_imshow(shift_img)
```

```python
cap = cv2.VideoCapture('/content/tinyv.mp4')
while cap.isOpened():
    ret, image = cap.read()

    if not ret:
      break

    cv2_imshow(image) # Note cv2_imshow, not cv2.imshow

    cv2.waitKey(1) & 0xff

cv2.destroyAllWindows()
cap.release()
```

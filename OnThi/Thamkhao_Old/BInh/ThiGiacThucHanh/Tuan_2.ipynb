```python
rows, cols = img.shape[:2]
plt.subplot(2,3,1)
plt.imshow(img)

plt.subplot(2,3,2)
M = np.float32([[1,0,500],
                    [0,1,0]])
dst = cv.warpAffine(img, M, (cols, rows))
plt.imshow(dst)

plt.subplot(2,3,3)
pts1 = np.float32([[0, 0],
                   [cols-1, 0], 
                   [0,rows-1]])
 
pts2 = np.float32([[cols-1,0],
                   [0, 0], 
                   [cols-1, rows-1]])
 
M = cv.getAffineTransform(pts1, pts2)
# print(M)
dst = cv.warpAffine(img, M, (cols, rows))
plt.imshow(dst)

plt.subplot(2,3,4)
M = np.float32([[2,0,0],
                    [0,1,0]])
dst = cv.warpAffine(img, M, (cols+1000, rows))
plt.imshow(dst)

plt.subplot(2,3,5)
M=cv.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv.warpAffine(img, M, (cols, rows))
plt.imshow(dst)

plt.subplot(2,3,6)
pts1 = np.float32([[0, 0],
                   [cols-1, 0], 
                   [0,rows-1]])
 
pts2 = np.float32([[500,0],
                   [cols+500, 0], 
                   [0,rows-1]])
 
M = cv.getAffineTransform(pts1, pts2)
# print(M)
dst = cv.warpAffine(img, M, (cols+1000, rows))
plt.imshow(dst)
```

```python
black=np.zeros((500,500,3),dtype=np.int8)
plt.imshow(black)
```

```python
myline=np.copy(black)
myrectangle=np.copy(black)
mytext=np.copy(black)
mycircle=np.copy(black)
myeclip=np.copy(black)
myall=np.copy(black)
cv.line(myline,(0,0),myline.shape[:-1],(255,0,0),3)
cv.rectangle(myrectangle,(200,200),(300,300),(0,255,0),3)
cv.putText(mytext,"Helloooooo",(100,350),1,2,(0,0,255),3)
cv.circle(mycircle,(250,250),50,(100,200,30),3)
cv.ellipse(myeclip,(250,250),(50,120),45,0,360.0,(30,30,30),5)
cv.line(myall,(150,150),(450,450),(0,255,0),3)
cv.rectangle(myall,(350,120),(450,250),(0,0,255),-1)
cv.circle(myall,(250,50),30,(100,200,10),3)
cv.circle(myall,(250,50),20,(100,200,10),-1)
cv.putText(myall,"Not funny shapes",(10,480),1,2,(255,0,0),3)
cv.ellipse(myall,(150,350),(40,90),45.0,0.0,360.0,(123,123,123),3)
cv.ellipse(myall,(150,350),(30,60),45.0,0.0,90.0,(123,123,123),-1)
cv.ellipse(myall,(150,350),(30,60),45.0,180.0,270.0,(123,123,123),-1)
plt.subplot(2,3,1)
plt.imshow(myline)
plt.subplot(2,3,2)
plt.imshow(myrectangle)
plt.subplot(2,3,3)
plt.imshow(mytext)
plt.subplot(2,3,4)
plt.imshow(mycircle)
plt.subplot(2,3,5)
plt.imshow(myeclip)
plt.subplot(2,3,6)
plt.imshow(myall)
```

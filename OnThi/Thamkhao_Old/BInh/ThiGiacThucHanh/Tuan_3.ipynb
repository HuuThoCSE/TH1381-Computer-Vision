```python
import cv2 as cv
import numpy as np
# hàm vẽ HÌNH TRÒN tại tọa độ x,y có bán ký là 10 khi bấm LMB(Left Mouse Button)
# khi bấm RMB(Right Mouse Button) sẽ vẽ HÌNH TRÒN khác màu khác
def draw_cirle(event,x,y,flags,param):
    if event==cv.EVENT_LBUTTONDOWN:
        cv.circle(img,(x,y,),10,(0,255,0),-1)
    if event==cv.EVENT_RBUTTONDOWN:
        cv.circle(img,(x,y,),10,(0,0,255),-1)

#tạo ma trận vuông có giá trị 0, là ma trận số nguyên dài,rộng 512, 3 kênh
img=np.zeros((512,512,3),np.uint8)

#tạo cửa sổ tên 'drawing'
cv.namedWindow(winname='drawing')
#tạo mouse callback cho cửa sổ
#mouse callback là loại callback khi có trỏ chuột tác động lên cửa sổ khai báo
cv.setMouseCallback('drawing',draw_cirle)
while True: #tạo vòng lập vô tận
    # hiển thị ảnh img với tên cửa sổ 'drawing'
    cv.imshow('drawing',img)
    # cv.waitKey(20) là đợi 20ms nếu không có key ra -1(tức false) và ngược lại
    # 0xFF==27 đợi nếu key đó là 27 tức là ESC thì true và ngược lại
    # cả 2 đk trên đúng thì ngưng hiện ảnh
    if cv.waitKey(20) & 0xFF==27:
        break
# đóng tất cả các cửa số nếu có nhằm giải phóng bộ nhớ?
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
# khai báo có đang nhấn hay không
drawing = False
# kb tọa độ x,y góc
ix,iy=-1,-1
#tạo hàm vẽ hình chữ nhật khi có sự kiện từ mouse
def draw_rectangle(event,x,y,flags,param):
    global ix,iy,drawing,mode
    if event==cv.EVENT_LBUTTONDOWN:
        # khi bấm chuột trái sẽ xét là nhấn và ghi lại tọa độ gốc tại lúc nhấn
        drawing=True
        ix,iy=x,y
    if event==cv.EVENT_MOUSEMOVE:
        # chạy khối lện khi con trỏ chuột di chuyển trên img
        if drawing == True:
            # nếu đg nhấn tức drawing thì sẽ vẽ hình CN từ tọa độ gốc tới
            # vị trí hiện tại của trỏ chuột
            cv.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
    elif event == cv.EVENT_LBUTTONUP:
        # chạy khối lệnh khi buôn chuột trái ra
        # đưa khóa có đang nhấn giữ về false
        drawing=False
        # vẽ ra hình chữ nhật tại điểm đầu đến vị trí hiện tại của trỏ chuột
        cv.rectangle(img,(ix,iy),(x,y),(0,0,255),-1)

#tạo ma trận vuông có giá trị 0, là ma trận số nguyên dài,rộng 512, 3 kênh
img=np.zeros((512,512,3),np.uint8)

#tạo cửa sổ tên 'drawing'
cv.namedWindow(winname='drawing')
#tạo mouse callback cho cửa sổ
#mouse callback là loại callback khi có trỏ chuột tác động lên cửa sổ khai báo
cv.setMouseCallback('drawing',draw_rectangle)
while True: #tạo vòng lập vô tận
    # hiển thị ảnh img với tên cửa sổ 'drawing'
    cv.imshow('drawing',img)
    # cv.waitKey(20) là đợi 20ms nếu không có key ra -1(tức false) và ngược lại
    # 0xFF==27 đợi nếu key đó là 27 tức là ESC thì true và ngược lại
    # cả 2 đk trên đúng thì ngưng hiện ảnh
    if cv.waitKey(20) & 0xFF==27:
        break
# đóng tất cả các cửa số nếu có nhằm giải phóng bộ nhớ?
cv.destroyAllWindows()
```

```python
img1=cv.imread('dog_backpack.png')
img2=cv.imread('watermark_no_copy.png')
img1=cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img2=cv.cvtColor(img2,cv.COLOR_BGR2RGB)
img2=cv.resize(img2,(300,400))
```

```python
img2gray=cv.cvtColor(img2,cv.COLOR_RGB2GRAY)
mymask=cv.bitwise_not(img2gray)
wtmask=cv.bitwise_or(img2,img2,mask=mymask)
plt.imshow(wtmask)
```

```python
# Góc trên bên trái
roi=img1[0:img2.shape[0],0:img2.shape[1]]
roi=cv.bitwise_or(roi,wtmask)
img1[0:img2.shape[0],0:img2.shape[1]]=roi
plt.imshow(roi)
```

```python
# Góc dưới bên trái
roi=img1[img1.shape[0]-img2.shape[0]:img1.shape[0],0:img2.shape[1]]
roi=cv.bitwise_or(roi,wtmask)
img1[img1.shape[0]-img2.shape[0]:img1.shape[0],0:img2.shape[1]]=roi
plt.imshow(roi)
```

```python
# Góc trên bên phải
roi=img1[0:img2.shape[0],img1.shape[1]-img2.shape[1]:img1.shape[1]]
roi=cv.bitwise_or(roi,wtmask)
img1[0:img2.shape[0],img1.shape[1]-img2.shape[1]:img1.shape[1]]=roi
plt.imshow(roi)
```

```python
# Góc dưới bên phải
roi=img1[img1.shape[0]-img2.shape[0]:img1.shape[0],img1.shape[1]-img2.shape[1]:img1.shape[1]]
roi=cv.bitwise_or(roi,wtmask)
img1[img1.shape[0]-img2.shape[0]:img1.shape[0],img1.shape[1]-img2.shape[1]:img1.shape[1]]=roi
plt.imshow(roi)
```
